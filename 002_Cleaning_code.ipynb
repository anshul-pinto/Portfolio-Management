```python
import pandas as pd
import numpy as np
```

```python
def get_empty_columns_names(df):
    s1 = df.isna().sum()
    vals = s1.values
    indexes = s1.index
    
    threshold = df.shape[0]/2
    
    return_list =[]
    
    for i in range(len(s1)):
        if s1[i]>threshold:
            return_list.append(indexes[i])
    
    return return_list
```

```python
def clean_the_df(df):
    df['Quarter end'] = pd.to_datetime(df['Quarter end'])
    df = df.sort_values(by=['Quarter end'])
    
    '''

    This is done to replace all 'None' Values in the dataframe with 
    the previous valid value

    '''

    for column in df.columns:
        df[column] = df[column].replace('None', value=np.nan)
    
    df = df.drop(columns = ['Price high', 'Price low','Shares', 'Shares split adjusted', 'Split factor',
                       ])
    df = df.set_index('Quarter end')
    
    for column in df.columns:
        df[column] = df[column].astype('float64')
    
    new_df = df.pct_change()
    
    to_remove = ['Non-controlling interest','Preferred equity','Goodwill & intangibles','Long-term debt','Dividend per share','Dividend payout ratio',
                 'Long-term debt to equity ratio','Free cash flow per share','Capital expenditures']
                #'Current Assets', 'Current Liabilities','Current ratio']

    new_df = new_df.drop(columns=to_remove)

    #names = get_empty_columns_names(new_df)

    #new_df = new_df.drop(columns=names)


    new_df = new_df.fillna(method='bfill')

    return new_df
```
